```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler(feature_range = (0 , 1))
```

```python
X = sc.fit_transform(X)
```

```python
from minisom import MiniSom
```

```python
ms = MiniSom(x = 10 , y = 10 , input_len = 15 , sigma = 1.0 , learning_rate = 0.5)
ms.random_weights_init(X)
ms.train_random(data = X , num_iteration = 100)
```

Initialising window that will contain the map

distance_mean() returns all interneuron distances in the map for all winning nodes , for getting o/p in the right order we have to get the transpose

highes mean interneural dist (mid) correspond to colour white and decrease gradually to other colours and smaller mids correspond to darker colours .

the outliers( with high mids ) are potential frauds .

red-circle : if customer didnt get approval
green-square : if customer got approval

```python
from pylab import bone , pcolor , colorbar , plot , show
bone()
pcolor(ms.distance_map().T)
colorbar()
markers = ['o', 's']
colours = ['r', 'g']
for i, x in enumerate(X):
    w = ms.winner(x)
    plot(w[0] + 0.5 ,
         w[1] + 0.5 ,
         markers[y[i]],
         markeredgecolor = colours[y[i]],
         markerfacecolor = 'None',
         markersize = 10,
         markeredgewidth = 2)
show()

```

```python
mappings = ms.win_map(X)
frauds = np.concatenate((mappings[(6,8)], mappings[(3,2)] ) , axis = 0 )
frauds = sc.inverse_transform(frauds)
```

# Predicting the probability of a fraud

matrix of features

```python
customers = data.iloc[:,1:].values
```

```python
customers
```

dependent variable

```python
fraud_ids = np.zeros(len(data))
for i in range(len(data)) :
  if data.iloc[i , 0 ] in frauds :
    fraud_ids[i] = 1
```

```python
for i in range(len(fraud_ids)) :
  if fraud_ids[i] == 1.0 :
    print(i)
```

# Neural Network

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
customers = std.fit_transform(customers)
```

```python
customers
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
classifier = Sequential()
classifier.add(Dense(units = 2, kernel_initializer = "uniform", activation = "relu" , input_dim = 15))
```

```python
classifier.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
classifier.fit(customers, fraud_ids , batch_size = 1 , epochs = 10)
```

```python
y_pred = classifier.predict(customers)
```

```python
y_pred
```

```python
y_pred = classifier.predict(customers)
y_pred = np.concatenate((data.iloc[:, 0:1].values, y_pred), axis = 1)
y_pred = y_pred[y_pred[:, 1].argsort()]
```
